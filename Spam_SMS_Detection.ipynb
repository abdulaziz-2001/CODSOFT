# EDA

**Importing the Required Dependencies**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

**Loading the Data**

```python
raw_sms_df = pd.read_csv('/content/spam.csv', encoding = 'latin1')
raw_sms_df
```

**Dropping off the Unwanted Columns**

```python
sms_df = raw_sms_df.drop(raw_sms_df.columns[[2,3,4]],axis = 1)
sms_df
```

**Renaming the Columns**

```python
sms_df.columns = ['class','text']
sms_df
```

```python
sms_df['class'] = sms_df['class'].map({'ham' : 1, 'spam' : 0})
sms_df
```

```python
sms_df['text'] = sms_df['text'].str.lower()
sms_df
```

**Separating the Feature(X) and Target(Y) Columns**

```python
X = sms_df['text']
Y = sms_df['class']
```

```python
X
```

```python
Y
```

**Splitting the Data into Train and Test Data**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Feature Extraction

```python
feature_extraction = TfidfVectorizer(min_df = 1 , stop_words = 'english')

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)
```

```python
print(X_test_features)
```

# Training and Evaluating the Model

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features,Y_train)
```

```python
# Prediction on test Data

te_data = model.predict(X_test_features)
accuracy_te_data = accuracy_score(Y_test,te_data)
report = classification_report(Y_test, te_data)
```

```python
print('Accuracy of the Test data : {:%}'.format( accuracy_te_data))
print('Classification Report :\n', report)
```

# Building a Prediction System

```python

input_sms =['Go until jurong point, crazy.. Available only in bugis n great world la e buffet... Cine there got a...']

input_da_features = feature_extraction.transform(input_sms)

# Classifying the SMS
classify = model.predict(input_da_features)
print(classify)

if classify[0] == 0:
  print('Spam')
else:
  print('Ham')
```
